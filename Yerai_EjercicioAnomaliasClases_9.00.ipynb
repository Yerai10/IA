1. Utiliza el Determinante de Covarianza Mínima para detectar los outliers y hacer predicciones con la misma. Calcula todas las métricas anteriores y muestra la matriz de confusión

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn.covariance import EllipticEnvelope

df = pd.read_csv("creditcard.csv")
df.dropna(axis=0, inplace=True)
df.drop(['Time'], axis=1, inplace=True)

# Separo las características (X) y la variable objetivo (y)
X_train, X_test, y_train, y_test = train_test_split(df.drop("Class", axis=1), df.Class, train_size=.7, random_state=42)

# Estandarizo las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entreno el modelo de Detección de Anomalías (Elliptic Envelope)
anomaly_detector = EllipticEnvelope(contamination=0.02)
anomaly_detector.fit(X_train_scaled)

```

```python
# Realizo predicciones en el conjunto de prueba
y_pred_anomaly = anomaly_detector.predict(X_test_scaled)

# Convierto las predicciones a etiquetas binarias (1 para normal, -1 para anomalía)
y_pred_anomaly_binary = [1 if pred == 1 else 0 for pred in y_pred_anomaly]


# Matriz de confusión
conf_anomaly = confusion_matrix(y_test, y_pred_anomaly_binary)
print("Matriz de Confusión:\n", conf_anomaly)


print("Accuracy: ", accuracy_score(y_test, y_pred_anomaly_binary))
print("Recall: ", recall_score(y_test, y_pred_anomaly_binary))
print("Precision: ", precision_score(y_test, y_pred_anomaly_binary))
print("Specificity: ", recall_score(y_test, y_pred_anomaly_binary, pos_label=0))
print("F1-score: ", f1_score(y_test, y_pred_anomaly_binary))

```

las metricas que he obtenido como resultado de la detección de anomalías utilizando el Determinante de Covarianza Mínima (MCD) indican un rendimiento bastante bajo en este conjunto de datos.

Accuracy (Exactitud): 0.0206

La exactitud es muy baja, lo que sugiere que el modelo tiene dificultades para clasificar correctamente las transacciones normales y fraudulentas.

Recall (Recuperación): 0.7941

La tasa de recuperación es relativamente alta, lo que indica que el modelo ha identificado correctamente casi el 80% de las transacciones fraudulentas.

Precision (Precisión): 0.0013

La precisión es extremadamente baja, indicando que la gran mayoría de las transacciones que el modelo clasifica como fraudulentas son falsos positivos.

Specificity (Especificidad): 0.0194

La especificidad es baja, indicando que el modelo tiene dificultades para identificar correctamente las transacciones normales.

F1-score: 0.0026

El F1-score, que combina precisión y recall, también es muy bajo.

Matriz de Confusión:

 [   57,  55]]

El modelo ha identificado 108 transacciones fraudulentas correctamente (verdaderos positivos) pero ha clasificado erróneamente 83655 transacciones normales como anomalías (falsos positivos).

Estos resultados sugieren que, aunque el modelo tiene un alto recall (buena capacidad para identificar transacciones fraudulentas), tiene dificultades para mantener la precisión y especificidad, clasificando muchas transacciones normales como anomalías (falsos positivos)

2. Utiliza el Isolation Forest para detectar los outliers y hacer predicciones con el mismo. Calcula las métricas y muestra la matriz de confusión

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

df = pd.read_csv("creditcard.csv")
df.dropna(axis=0, inplace=True)
df.drop(['Time'], axis=1, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(df.drop("Class", axis=1), df.Class, train_size=.7, random_state=42)

# Estandarizo las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entreno el modelo Isolation Forest
isolation_forest = IsolationForest(contamination=0.02, random_state=42)
isolation_forest.fit(X_train_scaled)

# Realizo predicciones en el conjunto de prueba
y_pred_isolation = isolation_forest.predict(X_test_scaled)

# Convierto las predicciones a etiquetas binarias (1 para normal, -1 para anomalía)
y_pred_isolation_binary = [1 if pred == 1 else 0 for pred in y_pred_isolation]

# Muestro la matriz de confusión
conf_isolation = confusion_matrix(y_test, y_pred_isolation_binary)
print("Confusion Matrix:\n", conf_isolation)

print("Accuracy: ", accuracy_score(y_test, y_pred_isolation_binary))
print("Recall: ", recall_score(y_test, y_pred_isolation_binary))
print("Precision: ", precision_score(y_test, y_pred_isolation_binary))
print("Specificity: ", recall_score(y_test, y_pred_isolation_binary, pos_label=0))
print("F1-score: ", f1_score(y_test, y_pred_isolation_binary))


```

6.Utiliza Logistic Regression sensible al coste y comprueba las métricas. ¿Ha mejorado el Recall? Imprime la matriz de confusión

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score

# Defino el modelo de Regresión Logística
model_lr_cost_sensitive = LogisticRegression(class_weight='balanced', random_state=42)

# Entreno el modelo
model_lr_cost_sensitive.fit(X_train_scaled, y_train)

# Realizo las predicciones en el conjunto de prueba
y_pred_lr_cost_sensitive = model_lr_cost_sensitive.predict(X_test_scaled)

# Calculo de las métricas
accuracy_lr_cost_sensitive = accuracy_score(y_test, y_pred_lr_cost_sensitive)
recall_lr_cost_sensitive = recall_score(y_test, y_pred_lr_cost_sensitive)
confusion_lr_cost_sensitive = confusion_matrix(y_test, y_pred_lr_cost_sensitive)

print(f'Accuracy del modelo de Regresión Logística sensible al coste: {accuracy_lr_cost_sensitive}')
print(f'Recall del modelo de Regresión Logística sensible al coste: {recall_lr_cost_sensitive}')
print(f'\nMatriz de Confusión:')
print(confusion_lr_cost_sensitive)
```

El recall anterior era de aproximadamente 22.79%, y después de aplicar la sensibilidad al costo, el recall ha mejorado considerablemente a aproximadamente 92.65%.

Se puede decir que el modelo sensible al costo y está identificando una proporción mucho mayor de las transacciones fraudulentas presentes en el conjunto de prueba, lo cual es un resultado positivo.

La matriz de confusión también muestra un aumento en los verdaderos positivos (transacciones fraudulentas correctamente identificadas) y una disminución en los falsos negativos.

7.Haz lo mismo haciendo undersampling

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score
from sklearn.preprocessing import StandardScaler

# Realizo undersampling en el conjunto de entrenamiento
undersampling_train = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = undersampling_train.fit_resample(X_train_scaled, y_train)

# Defino el modelo de Regresión Logística
model_lr_undersampling = LogisticRegression(class_weight='balanced', random_state=42)

# Entreno el modelo con el conjunto de entrenamiento undersampled
model_lr_undersampling.fit(X_train_resampled, y_train_resampled)

# Realizo las predicciones en el conjunto de prueba
y_pred_lr_undersampling = model_lr_undersampling.predict(X_test_scaled)

# Calculo todas las métricas
accuracy_lr_undersampling = accuracy_score(y_test, y_pred_lr_undersampling)
recall_lr_undersampling = recall_score(y_test, y_pred_lr_undersampling)
confusion_lr_undersampling = confusion_matrix(y_test, y_pred_lr_undersampling)

print(f'Accuracy del modelo de Regresión Logística con undersampling: {accuracy_lr_undersampling}')
print(f'Recall del modelo de Regresión Logística con undersampling: {recall_lr_undersampling}')
print(f'\nMatriz de Confusión:')
print(confusion_lr_undersampling)
```
